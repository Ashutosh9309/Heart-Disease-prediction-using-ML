# Data preprocessing

```python
X = heart.iloc[:,:-1]
y=heart['target']
```

```python
X.shape
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

# DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(X_train,y_train)
```

```python
prediction = dt.predict(X_test)
```

```python
prediction
```

```python
from sklearn import metrics
print('r2_score',metrics.r2_score(y_test, prediction) )
```

```python
print('Accuracy ', metrics.accuracy_score(y_test, prediction))
```

```python
print(metrics.confusion_matrix(y_test, prediction))
```

```python
print(metrics.classification_report(y_test, prediction))
```

```python
from sklearn.tree import export_graphviz
import pydot
from IPython.display import Image 
```

```python
export_graphviz(dt,out_file='decision_tree.dot',feature_names=list(heart.columns[:-1]),rounded=True, filled=True )

(graph,) = pydot.graph_from_dot_file('decision_tree.dot')
graph.write_png('decision_tree.png')
Image('decision_tree.png')

```

# Using RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(X_train, y_train)
```

```python
predict = rfc.predict(X_test)
```

```python
print(metrics.classification_report(y_test, predict))
```

```python
print(metrics.confusion_matrix(y_test, predict))
```

# Hyper Tunning model

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
params = [int(i) for i in np.linspace(start=10, stop=120, num=6)]
max_depth = [int(i) for i in np.linspace(start=1, stop=120, num=6)]
```

```python
random_grid = {
    'n_estimators':params,
    'max_features':['auto', 'sqrt'],
    'max_depth':max_depth,
    'min_samples_split':[5,10,15,100]
}
```

```python
rfc_random = RandomizedSearchCV(
    estimator=RandomForestClassifier(),
    param_distributions = random_grid,
    verbose = 2,
    cv = 3, n_jobs=-1

)
```

```python
rfc_random.fit(X_train, y_train)
```

```python
rfc_random.best_params_
```

```python
predict = rfc_random.predict(X_test)
```

```python
predict
```

```python
print(metrics.classification_report(y_test, predict))
```

```python
print(metrics.confusion_matrix(y_test, predict))
```

# Using KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=2)
```

```python
knn.fit(X_train, y_train)
```

```python
pre = knn.predict(X_test)
```

```python
pre
```

```python
print(metrics.classification_report(y_test, pre))
```

```python
print(metrics.confusion_matrix(y_test, pre))
```

```python
error_rate = []

for i in range(1,50):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    predictions = knn.predict(X_test)
    error_rate.append(np.mean(y_test!=predictions))
    
```

```python
plt.plot(range(1,50), error_rate, 'o-')
```

```python
knn = KNeighborsClassifier(n_neighbors=24)
```

```python
knn.fit(X_train, y_train)
```

```python
predict = knn.predict(X_test)
```

```python
print(metrics.classification_report(y_test, predict))
```
